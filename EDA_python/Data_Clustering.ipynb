```python
import sklearn.datasets as datasets
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate artificia; data
plt.jet()
X, Y = datasets.make_blobs(centers = 5, cluster_std = 0.5, random_state = 0)
plt.scatter(X[:,0],X[:,1]);
```

```python
plt.scatter(X[:,0],X[:,1], c=Y); # cluster data by colour
```

```python
# typically we dont know the information in Y, however we can revoer it from the data
# to do this we use kMeans algorithm
from sklearn.cluster import KMeans
kmeans = KMeans(4, random_state=8)
Y_hat = kmeans.fit(X).labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat);
```

```python
# if we're interested in finding the means for each cluster:
plt.scatter(X[:,0], X[:,1], c=Y_hat, alpha=0.5)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(Y_hat))
print(mu)
```
