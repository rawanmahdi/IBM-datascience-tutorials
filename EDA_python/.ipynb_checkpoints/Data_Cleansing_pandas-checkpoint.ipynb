Data cleansing process is divided into the below tasks
1. importing required libraries
2. loading data set from source and displaying 
3. removing unused columns, renaming labels
4. replacing values of rows to make them mmore meaningful

```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv("data.csv")
df.head(5)
```

```python
# define the columns to remove
drop_col = ['Engine Fuel Type', 
           'Engine HP',
           'Engine Cylinders',
           'MSRP']
df.drop(drop_col, inplace = True, axis = 1)
df.head(5)
```

```python
# define replacment names through dictionary
new_name = {'Make':"Company Name",
           'Year': 'Release Year'}
df.rename(columns = new_name, inplace = True)
df.head()
```

```python
# replacing values: df = df.replace({"col_name": {replacement dictionary}})
```
