```python
# make scatter plot of data
# extract features data
all_features = []
features = iris.data[:, [0,1,2,3]]
features.shape
```

```python
# extract targets data
targets = iris.target
targets.reshape(targets.shape[0],-1)
targets.shape
```

```python
# append observations to features list
for observ in features:
    all_features.append([observ[0]+observ[1]+observ[2]+observ[3]])
print(all_features)    
```

```python
import matplotlib.pyplot as plt
plt.scatter(all_features, targets, color = 'red', alpha = 1.0)
plt.rcParams['figure.figsize'] = [10,8]
plt.title('Iris Dataset Scatterplot')
plt.xlabel('Features')
plt.ylabel('Targets')
```

```python
# now we can try plotting the relationship between Sepal length and Sepal width

all_features = []
targets = []

for feature in features:
    all_features.append(feature[0]) # sepal length
    targets.append(feature[1]) # sepan width

categories = ('Setosa','Versicolor', 'Virginica')
colors = ('blue', 'green', 'red')
data = ((all_features[:50], targets[:50]), (all_features[50:100], targets[50:100]), (all_features[100:150],targets[100:150]))

for item, color, group in zip(data,colors,categories):
    x,y = item
    plt.scatter(x,y,color=color, alpha = 1)
    plt.title('Iris Dataset Scatterplot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# now lets plot correlation bw pedal length and peda; width
all_features = []
targets = []

for feature in features:
    all_features.append(feature[2]) # sepal length
    targets.append(feature[3]) # sepan width

categories = ('Setosa','Versicolor', 'Virginica')
colors = ('blue', 'green', 'red')
data = ((all_features[:50], targets[:50]), (all_features[50:100], targets[50:100]), (all_features[100:150],targets[100:150]))

for item, color, group in zip(data,colors,categories):
    x,y = item
    plt.scatter(x,y,color=color, alpha = 1)
    plt.title('Iris Dataset Scatterplot')
plt.xlabel('Pedal Length')
plt.ylabel('Pedal Width')
plt.show()
```

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

n_neighbors = 15

# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
for weights in ['uniform', 'distance']:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))

plt.show()
```
