```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv('data.csv')
df.head(5)
```

```python
df.tail(5)
```

```python
df.dtypes # outputs data types in data frame
```

```python
# drop the columns of data that we never user:
# .drop( list of labels we want to drop, axis number - 0 if dropping labels from index, 1 if dropping from column)
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis = 1)
```

```python
df.head(5) # observe new dataframe with removed columns 
```

```python
# we can also drop duplicate rows
df.shape
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count() # seehow removing duplicaet rows dropped count
```

```python
# dealing with missing values:
print(df.isnull().sum()) # isnull() returns boolean if a box contains no value
```

```python
df = df.dropna() # removing entries with null values
print(df.isnull().sum())
```

```python
# cleansing data: detecting outliers:

# use seaborn to view some plots:
sns.boxplot(x=df['Engine HP'])
```

```python
sns.boxplot(x=df['Engine Cylinders'])
```

```python
# to remove outliers:
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.shape
```

```python
# histograms - frequency of occurrrences of a variable in an interval
df.Make.value_counts().nlargest(40).plot(kind = 'bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make");
```

```python
# heatmaps - helps determine depndant variables and relationships between features
plt.figure(figsize=(10,5))
c = df.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```

```python
# scatterplot - find correlation bw two variables

fig, axis = plt.subplots(figsize=(10,5))
axis.scatter(df['Engine HP'], df["Year"])
axis.set_xlabel('Engine HP')
axis.set_ylabel('Year')
plt.show
```
